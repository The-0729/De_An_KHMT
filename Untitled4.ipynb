```python
colors = {
    'background': '#FFF9D4',
    'bodyColor':'#FFF9D4',
    'text': '#111111'
}
def get_page_heading_style():
    return {'backgroundColor': colors['background']}


def get_page_heading_title():
    return html.H1(children='Stock Dashboard',
                                        style={
                                        'textAlign': 'center',
                                        'color': colors['text']
                                    })

def get_page_heading_subtitle():
    return html.Div(children='Visualize stock data generated from sources all over Vietnam.',
                                         style={
                                             'textAlign':'center',
                                             'color':colors['text']
                                         })

def generate_page_header():
    main_header =  dbc.Row(
                            [
                                dbc.Col(get_page_heading_title(),md=12)
                            ],
                            align="center",
                            style=get_page_heading_style()
                        )
    subtitle_header = dbc.Row(
                            [
                                dbc.Col(get_page_heading_subtitle(),md=100)
                            ],
                            align="center", justify="center",
                            style=get_page_heading_style()
                        )
    header = (main_header,subtitle_header)
    return header
```

```python
def generate_card_content(card_header,card_value,overall_value):
    card_head_style = {'textAlign':'center','fontSize':'150%'}
    card_body_style = {'textAlign':'center','fontSize':'200%'}
    card_header = dbc.CardHeader(card_header,style=card_head_style)
    card_body = dbc.CardBody(
        [
            html.H5(f"{card_value:,}", className="card-title",style=card_body_style),
            html.P(
                "Ticker: {}".format(overall_value),
                className="card-text",style={'textAlign':'center'}
            ),
        ]
    )
    card = [card_header,card_body]
    return card
```

```python
def generate_cards():
    df = df_hnx_all[df_hnx_all['Date'] == now].sort_values('Volume', ascending = 0)
    cards = html.Div(
        [
            dbc.Row(
                [
                    dbc.Col(dbc.Card(generate_card_content("2nd Volume",df.iloc[1]['Volume'],df.iloc[1]['Ticker']),color="grey", inverse=True),md=dict(size=4,offset=0)),
                    dbc.Col(dbc.Card(generate_card_content("1st Volume",df.iloc[0]['Volume'],df.iloc[0]['Ticker']),color="#F8FF36", inverse=True),md=dict(size=4)),
                    dbc.Col(dbc.Card(generate_card_content("3rd Volume",df.iloc[2]['Volume'],df.iloc[2]['Ticker']),color="warning", inverse=True),md=dict(size=4))
                ],
                className="mb-4",
            ),
        ],id='card'
    )
    return cards
```

```python
dff = df_hnx_all[df_hnx_all['Date'] > '2015-01-01']
dff
```

```python
list_percent = dff['Ticker'].unique()
list_percent
```

```python
def percent(list_percent):
    list_card = []
    for i in list_percent:
        b = dff[dff['Ticker'] == i][['Close']]
        list_card.append(round((b.iloc[0][0]/b.iloc[1][0] - 1) * 100,2))
    return list_card
```

```python
x = percent(list_percent)
x
```

```python
percentile_list1 = pd.DataFrame(
    {'Ticker': list_percent,
     
     'Change %': x,
    })
percentile_list2 = pd.DataFrame(
    {'Ticker': list_percent,
     'Change %': x,
    })
```

```python
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = html.Div([
       
        dbc.Container([
            html.Div([
                page_header[0],
                page_header[1]], style={'backgroundColor': colors['bodyColor'], "border":"2px black solid"}),
            
            html.Hr(),
            
            dbc.Row(
                [
                    dbc.Col(html.H1("Top Volume Yesterday"), style={'textAlign': 'center'}),       
                ]        
            ),
            
            generate_cards(),
            
            generate_cards3(),

            html.Hr(),
            
            dbc.Row(
                [
                    dbc.Col(html.H1("Top Increase Yesterday"), style={'textAlign': 'center'}),       
                ]        
            ),
            
            generate_cards1(),
            
            html.Hr(),
            
            dbc.Row(
                [
                    dbc.Col(html.H1("Top Decrease Yesterday"), style={'textAlign': 'center'}),       
                ]        
            ),
            
            generate_cards2(),
            
            html.Hr(),
            
            dbc.Row(
                [
                    dbc.Col(html.H1("Select Ticker")),       
                ]         
            ),
            
            dbc.Row(
                [
                    dbc.Col(dcc.Dropdown(
                    id='ticker_list1',
                    options=[{'label': item, 'value': item} for item in hnx_ticker_list],
                    value='AAV',style={'width': '40%'}
                    ))
                ]         
            ),
            
            dbc.Row(
                [
                    dbc.Col(html.H5("Single chart of price of each ticker"), style={'textAlign': 'center'}),       
                ], align="center", justify="center",         
            ),
            
            html.Div(dcc.Graph(id='graph1')
            ,style={'width': '100%', 'display': 'inline-block', "border":"2px black solid", 'margin':25, 'margin-left':'0px'}),
            
            #sheet 2
            html.Hr(),
            dbc.Row(
                [
                    dbc.Col(html.H1("Select Ticker")),       
                ]         
            ),
            
            dbc.Row(
                [
                    dbc.Col(dcc.Dropdown(
                    id='ticker_list2',
                    options=[{'label': item, 'value': item} for item in hnx_ticker_list],
                    value='AAV',style={'width': '40%'}
                    ))
                ]         
            ),
            
            dbc.Row(
                [
                    dbc.Col(html.H5("Single chart of volumn of each ticker"), style={'textAlign': 'center'}),       
                ], align="center", justify="center",         
            ),
            
            html.Div(dcc.Graph(id='graph2')
            ,style={'width': '100%', 'display': 'inline-block', "border":"2px black solid", 'margin':25, 'margin-left':'0px'}),
            
            # sheet 3
            html.Hr(),
            
            dbc.Row(
                [
                dbc.Col(html.H4("Ticker 1"), width=2),
                dbc.Col(html.H4("Ticker 2"), width=2),
                dbc.Col(html.H4("Ticker 3"), width=2),
                ], justify="around"
            ),
            
            dbc.Row(
                [
                    dbc.Col(dcc.Dropdown(
                    id='ticker_list3',
                    options=[{'label': item, 'value': item} for item in hnx_ticker_list],
                    value='AAV',style={'width': '80%'}
                    ),width=2),
                    
                    dbc.Col(dcc.Dropdown(
                    id='ticker_list4',
                    options=[{'label': item, 'value': item} for item in hnx_ticker_list],
                    value='ACM',style={'width': '80%'}
                    ),width=2),
                    
                    dbc.Col(dcc.Dropdown(
                    id='ticker_list5',
                    options=[{'label': item, 'value': item} for item in hnx_ticker_list],
                    value='ADC',style={'width': '80%'}
                    ),width=2)
                ], justify="around",         
            ),
            
            dbc.Row(
                [
                    dbc.Col(html.H5("Multi line chart of volumn of selected ticker"), style={'textAlign': 'center'}),       
                ], align="center", justify="center",         
            ),
            
            html.Div(dcc.Graph(id='graph3')
            ,style={'width': '100%', 'display': 'inline-block', "border":"2px black solid", 'margin':25, 'margin-left':'0px'}),
        ])
])

@app.callback([
    Output('graph1', 'figure'),
    Output('card','children'),
    Output('card1','children')],
    Input('ticker_list1', 'value'))
def set_market_options(selected_ticker):       
    fig = go.Figure(data=[go.Candlestick(x=df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]['Date'],
                    open=df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]['Open'],
                    high=df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]['High'],
                    low=df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]['Low'],
                    close=df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]['Close'], showlegend=False)])
    return fig, generate_cards(), generate_cards1()

@app.callback(
    Output('graph2', 'figure'),
    Input('ticker_list2', 'value'))
    
def set_market_options1(selected_ticker):
    df = df_hnx_all[df_hnx_all['Ticker'] == selected_ticker]
    fig = px.line(df, x='Date', y='Volume')
    return fig

@app.callback(
    Output('graph3', 'figure'),
    [
        Input('ticker_list3', 'value'),
        Input('ticker_list4', 'value'),
        Input('ticker_list5', 'value')
    ]
    )
def set_market_options1(selected_ticker1, selected_ticker2, selected_ticker3):
    df1 = df_hnx_all[df_hnx_all['Ticker'] == selected_ticker1][['Date','Close']]
    df2 = df_hnx_all[df_hnx_all['Ticker'] == selected_ticker2][['Date','Close']]
    df3 = df_hnx_all[df_hnx_all['Ticker'] == selected_ticker3][['Date','Close']]

    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df1['Date'], y=df1['Close'], name=selected_ticker1,
                    line_shape='linear'))
    
    fig.add_trace(go.Scatter(x=df2['Date'], y=df2['Close'], name=selected_ticker2,
                    line_shape='linear'))
    
    fig.add_trace(go.Scatter(x=df3['Date'], y=df3['Close'], name=selected_ticker3,
                    line_shape='linear'))
    fig.update_layout(legend=dict(y=0.5, traceorder='reversed', font_size=16))
    fig.update_layout(legend=dict(y=0.5, traceorder='reversed', font_size=16))
    fig.update_layout(legend=dict(y=0.5, traceorder='reversed', font_size=16))
    return fig

if __name__ == '__main__':
    app.run_server(debug=False)
```
